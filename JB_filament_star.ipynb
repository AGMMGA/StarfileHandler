```python
import re
import copy

import pandas as pd
import numpy as np

from pathlib import Path

from star_parser import StarParser, StarTabDf
```

```python
work_folder = Path(r"/mnt/DATA/jitske/JB_20210825_FT040_NeCeN/csparc2relion")
star_in = work_folder/"P101_J25_passthrough_particles_class_2.star"
star_in.exists()
```

```python
parser = StarParser(star_in)
data = parser.parse()
```

```python
ptcls = data["data_particles"]
ptcls.rename_columns(['AnglePsi'], ['AnglePsiPrior'], store=True)
```

```python
column = ptcls.to_df()['HelicalTubeID']
keys = sorted(list(set(column)))
new_ids = range(1,len(keys)+1)
mapping = dict(zip(keys, [str(i) for i in new_ids]))
new_columns = []
for i in column:
    new_columns.append(mapping[str(i)])
```

```python
ptcls.fill_column('ClassNumber', '1', overwrite=True, store=True)
```

```python
ptcls.remove_columns(['HelicalTubeID'], store=True)
newc = pd.DataFrame({'HelicalTubeID':new_columns})
ptcls.add_columns(newc, store=True)
```

```python
ptcls.fill_column('AnglePsiFlipRatio', '0.5', create=True, store=True)
```

```python
#HelicalTrackLengthA
ids = ptcls.to_df()['HelicalTubeID']
counting_dict = dict(zip(ids, [0]*len(ids)))
track_lengths = []
for i in ids:
    new_count = counting_dict[i] +1
    track_lengths.append(str((82*new_count)))
    counting_dict[i] = new_count
print(type(track_lengths[0]))
```

```python
track_df = pd.DataFrame({'HelicalTrackLengthAngst':track_lengths})
ptcls.add_columns(track_df,store=True)
```

```python
mic_names = ptcls.to_df()['MicrographName']
mcor_jobs = {'J2/motioncorrected/': 'MotionCorr/job106',
"J4/motioncorrected/": "MotionCorr/job108"}
new_names = []
pattern_length = len('J4/motioncorrected/000000555374200640504_')
for m in mic_names:
    m = m.replace('_patch_aligned_doseweighted.mrc', '.mrc')
    m = m[pattern_length:]
    assert m.startswith('FoilHole')
    assert m.endswith('.mrc')
    new_names.append(m)
renamed_df = pd.DataFrame({'MicrographName':new_names})
ptcls.remove_columns(['MicrographName'], store=True)
ptcls.add_columns(renamed_df,store=True)
```

```python
img_names = ptcls.to_df()['ImageName']
new_img_names = []
pattern = re.compile('J[0-9]{2}\/[a-z]+\/[0-9]+_')
replacement = 'Extract/job255/Micrographs/'
for i in img_names:
    i = re.sub(pattern, replacement, i)
    i = i.replace('_patch_aligned_doseweighted_particles.mrc', '.mrcs')
    assert i.endswith('.mrcs')
    new_img_names.append(i)
new_img_names
df = pd.DataFrame({'ImageName': new_img_names}, dtype=str)
ptcls.remove_columns(['ImageName'], store=True)
ptcls.add_columns(df, store=True)
```

```python
old_order = list(ptcls.to_df().columns)
new_order = old_order[-2:] + old_order[:-2]
ptcls.reorder_columns(new_order, store=True)
```

```python
new_star = work_folder/'J25_passthrough_particles_class_2_modified.star'
with open(new_star, 'w') as f:
    f.writelines(ptcls.to_star())
```
